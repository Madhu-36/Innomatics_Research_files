```python
import sqlite3
import pandas as pd

# Read the SQL script from the file
with open('/content/restaurants.sql', 'r') as f:
    sql_script = f.read()

conn = sqlite3.connect(':memory:')
conn.executescript(sql_script)

tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)

if not tables.empty:

    table_name = tables.iloc[0, 0]
    rst = pd.read_sql(f"SELECT * FROM {table_name}", conn)

    output_filename = 'restaurants.xlsx'
    rst.to_excel(output_filename, index=False)
    print(f"file successfully completed: {output_filename}")
else:
    print("Error: No tables were found in the provided SQL script.")
```

```python
# Group by cuisine and calculate the mean of total_amount
avg_order_value_by_cuisine = final_dataset.groupby('cuisine')['total_amount'].mean().reset_index()

# Find the cuisine with the highest average order value
highest_avg_cuisine = avg_order_value_by_cuisine.loc[avg_order_value_by_cuisine['total_amount'].idxmax()]

print(f"The cuisine with the highest average order value is {highest_avg_cuisine['cuisine']} with an average of {highest_avg_cuisine['total_amount']:.2f}")

# Display the results sorted by average order value
display(avg_order_value_by_cuisine.sort_values(by='total_amount', ascending=False))
```

```python
# Group by user_id and sum the total_amount
user_total_spending = final_dataset.groupby('user_id')['total_amount'].sum().reset_index()

# Filter users with total spending > 1000
users_above_1000 = user_total_spending[user_total_spending['total_amount'] > 1000]

# Count the number of distinct users
distinct_user_count = users_above_1000['user_id'].nunique()

print(f"Number of distinct users with total spending > 1000: {distinct_user_count}")
```

```python
import pandas as pd

# Define the bins and labels for the rating ranges
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0 – 3.5', '3.6 – 4.0', '4.1 – 4.5', '4.6 – 5.0']

# Create a new column for the rating range
final_dataset['rating_range'] = pd.cut(final_dataset['rating'], bins=bins, labels=labels, include_lowest=True)

# Group by the rating range and sum the total_amount
# Added observed=False to resolve the FutureWarning
revenue_by_rating = final_dataset.groupby('rating_range', observed=False)['total_amount'].sum().reset_index()

# Find the range with the highest revenue
highest_revenue_range = revenue_by_rating.loc[revenue_by_rating['total_amount'].idxmax()]

print(f"The rating range with the highest total revenue is {highest_revenue_range['rating_range']} with a total of {highest_revenue_range['total_amount']:.2f}")

# Display the comparison
display(revenue_by_rating.sort_values(by='total_amount', ascending=False))
```

```python
# Filter for Gold members
gold_members_df = final_dataset[final_dataset['membership'] == 'Gold']

# Group by city and calculate the mean of total_amount
avg_revenue_city = gold_members_df.groupby('city')['total_amount'].mean().reset_index()

# Find the city with the highest average revenue
highest_avg_city = avg_revenue_city.loc[avg_revenue_city['total_amount'].idxmax()]

print(f"Among Gold members, the city with the highest average order value is {highest_avg_city['city']} with an average of {highest_avg_city['total_amount']:.2f}")

# Display the results sorted by average order value
display(avg_revenue_city.sort_values(by='total_amount', ascending=False))
```

```python
# Calculate number of distinct restaurants and total revenue per cuisine
cuisine_stats = final_dataset.groupby('cuisine').agg(
    distinct_restaurants=('restaurant_id', 'nunique'),
    total_revenue=('total_amount', 'sum')
).reset_index()

# Display the results
display(cuisine_stats.sort_values(by='distinct_restaurants'))
```

```python
# Calculate the total number of orders
total_orders = len(final_dataset)

# Calculate the number of orders placed by Gold members
gold_orders_count = len(final_dataset[final_dataset['membership'] == 'Gold'])

# Calculate the percentage
gold_percentage = (gold_orders_count / total_orders) * 100

print(f"Total orders: {total_orders}")
print(f"Gold member orders: {gold_orders_count}")
print(f"Percentage of orders by Gold members: {gold_percentage:.2f}%")
print(f"Rounded percentage: {round(gold_percentage)}%")
```

```python
# Group by restaurant (using ID and Name) to calculate stats
res_stats = final_dataset.groupby(['restaurant_id', 'restaurant_name_x']).agg(
    order_count=('order_id', 'count'),
    avg_order_value=('total_amount', 'mean')
).reset_index()

# Filter for restaurants with less than 20 orders
filtered_res = res_stats[res_stats['order_count'] < 20]

# Sort by average order value descending and display the top results
top_restaurants = filtered_res.sort_values(by='avg_order_value', ascending=False)
display(top_restaurants.head(10))
```

```python
options = ['Grand Cafe Punjabi', 'Grand Restaurant South Indian', 'Ruchi Mess Multicuisine', 'Ruchi Foods Chinese']

# Filter the restaurant stats for the specific options
selected_stats = res_stats[res_stats['restaurant_name_x'].isin(options)]

# Sort by average order value
display(selected_stats.sort_values(by='avg_order_value', ascending=False))
```

```python
# Group by membership and cuisine and sum the total_amount
revenue_by_combo = final_dataset.groupby(['membership', 'cuisine'])['total_amount'].sum().reset_index()

# Filter for the specific combinations mentioned in the options
options_list = [
    ('Gold', 'Indian'),
    ('Gold', 'Italian'),
    ('Regular', 'Indian'),
    ('Regular', 'Chinese')
]

# Create a boolean mask for filtering
mask = revenue_by_combo.apply(lambda x: (x['membership'], x['cuisine']) in options_list, axis=1)
filtered_combos = revenue_by_combo[mask]

# Display the results sorted by revenue
display(filtered_combos.sort_values(by='total_amount', ascending=False))
```

```python
# Convert order_date to datetime format
final_dataset['order_date'] = pd.to_datetime(final_dataset['order_date'], dayfirst=True)

# Extract the quarter
final_dataset['quarter'] = final_dataset['order_date'].dt.to_period('Q')

# Group by quarter and sum total_amount
quarterly_revenue = final_dataset.groupby('quarter')['total_amount'].sum().reset_index()

# Find the quarter with the highest revenue
highest_revenue_quarter = quarterly_revenue.loc[quarterly_revenue['total_amount'].idxmax()]

print(f"The quarter with the highest total revenue is {highest_revenue_quarter['quarter']} with a total of {highest_revenue_quarter['total_amount']:.2f}")

# Display the quarterly breakdown
display(quarterly_revenue.sort_values(by='total_amount', ascending=False))
```
